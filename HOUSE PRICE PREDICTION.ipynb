```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
import math
#function to calculate 
def y_predicted(x):
    ret_val=model.intercept_+model.coef_*x 
    ret_val=np.round(ret_val,2) 
    return ret_val
#read csv file
file=pd.read_csv('C:\\Users\\HP\\Downloads\\house-prices.csv')
file.head (5)

```

```python
#convert row matrix into column matrix
y=file['Price']
x=file['size']
x=x.array.reshape(len(x),1) 
y=y.array.reshape(len(x),1)
```

```python
#make model
model=LinearRegression() 
model.fit(x,y)
print("Theta Zero or Intercept : ",model.intercept_) 
print("Theta One or Slope: ",model.coef_)
```

```python
#find y predicted 
y_p=y_predicted(x)
print("Y Predicted : \n",y_p, "\n")
```

```python
#plot graph to show the actual y and predicted y
plt.scatter (x,y,color='purple')
plt.scatter (x,y_p, color='brown')
plt.plot(x,y_p, color='black')
plt.xlabel('Size of the house')
plt.ylabel('Price of the house')
plt.title(" Linear Regression in one variable ")
plt.legend (['Hypothesis line', 'Original Values', 'Predicted values'])
plt.show()
```

```python
#predict value for a new data 
new_x=int(input("Enter the size of the house: ")) 
new_predicted_y=y_predicted (new_x) 
print("Predicted price for entered size is : ",new_predicted_y)
```
